```python
#Declarar una variables
resCorrec_int: 0
resIncorrec_int: 0
resBlanco_int: 0
totalRes_int: 0

```

```python
#entradas
resCorrec_int= int(input("Ingrese Cantidad de respuestas correctas:"))
resIncorrec_int= int(input("Ingrese Cantidad de respuestas incorrectas:"))
resBlanco_int= int(input("Ingrese Cantidad de respuestas blanco:"))
```

```python
#proceso
totalRes_int = (resCorrec_int*4) + (resIncorrec_int*(-1)) + (resBlanco_int*0) / 3
```

```python
#salidas
print("Este es su total de respuestas" ,totalRes_int)
```
