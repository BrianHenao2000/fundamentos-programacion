```python
#Construir una funcion que reciba como parametro un entero y retorne el caracter al cual pertenece ese entero como codigo ASCII...
def buscar_car_Ascii(palabra,letra):
  Ascii=palabra.find(letra)
  return Ascii

palabra_one=input("Ingrese un numero: ")
letra_opc=input("Ingrese a cual caracter(número) le desea saber la posición: ")

busqueda = buscar_car_Ascii(palabra_one,letra_opc)
print("posicion: " ,str(busqueda))
if busqueda >=0:
  print("posicion de letra : ",str(busqueda))
else:
  print("El texto buscado no existe en la cadena")
```
